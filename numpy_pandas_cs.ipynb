```python
data = np.genfromtxt(
    r"C:\Users\ASUS\Downloads\auto-mpg.csv",
    delimiter=",",
    skip_header=1,
    dtype=str
)

# 3. Compute the 25th, 50th, and 75th percentiles of the weight column using NumPy.

numeric = data[:, :-1]              # All columns except last
car_names = data[:, -1]             # Last column

numeric[numeric == "?"] = "nan"

numeric = numeric.astype(float)

weight = numeric[:, 4]

p25 = np.nanpercentile(weight, 25)
p50 = np.nanpercentile(weight, 50)
p75 = np.nanpercentile(weight, 75)

print("25th Percentile of Weight:", p25)
print("50th Percentile (Median) of Weight:", p50)
print("75th Percentile of Weight:", p75)
```

```python
# 4. Convert the acceleration column into a NumPy array and normalize its values(scale between 0 and 1).
acceleration = numeric[:, 5]

# Convert to float
acceleration = acceleration.astype(float)

# Min-max normalization: scaled = (x - min) / (max - min)
acc_min = np.nanmin(acceleration)
acc_max = np.nanmax(acceleration)

acc_normalized = (acceleration - acc_min) / (acc_max - acc_min)

print("Normalized Acceleration Values:")
print(acc_normalized)
```

```python
# 5. Increase all horsepower values by 10% and store the updated values in a newNumPy array. Handle missing data (if any) by replacing it with the mean of the column before applying the increase.
horsepower = numeric[:, 3].astype(float)

hp_mean = np.nanmean(horsepower)
horsepower_filled = np.where(np.isnan(horsepower), hp_mean, horsepower)

horsepower_updated = horsepower_filled * 1.10

print("Original Horsepower (first 10):", horsepower[:10])
print("Horsepower After 10% Increase (first 10):", horsepower_updated[:10])
```

```python
# 6. Find the average displacement of cars with an origin of 2 (Europe) using NumPy indexing.
displacement = numeric[:, 2].astype(float)
origin = numeric[:, 7].astype(float)

european_cars = displacement[origin == 2]

avg_displacement = np.nanmean(european_cars)

print("Average displacement of European cars:", avg_displacement)
```

```python
#7. Matrix Operations

mpg = numeric[:, 0].astype(float)
hp = numeric[:, 3].astype(float)
weight = numeric[:, 4].astype(float)

hp_mean = np.nanmean(hp)
hp = np.where(np.isnan(hp), hp_mean, hp)

X = np.column_stack((mpg, hp, weight))

v = np.array([1, 0.5, -0.2])

dot_result = np.dot(X, v)

print("Dot product result (first 10 values):")
print(dot_result[:10])
```

```python
# 8. Sorting
model_year = numeric[:, 6].astype(float)

sorted_indices = np.argsort(model_year)[::-1]

top_5_cars = car_names[sorted_indices][:5]

print("Top 5 cars by model year:")
for car in top_5_cars:
    print(car)
```

```python
# 9. Correlation
mpg = numeric[:, 0].astype(float)
weight = numeric[:, 4].astype(float)

correlation_matrix = np.corrcoef(mpg, weight)
pearson_corr = correlation_matrix[0, 1]

print("Pearson Correlation between MPG and Weight:", pearson_corr)
```

```python
# 10.Conditional Aggregates
mpg = numeric[:, 0].astype(float)
cylinders = numeric[:, 1].astype(float)

unique_cyl = np.unique(cylinders)

print("Mean MPG by Cylinder Count:\n")
for cyl in unique_cyl:
    mean_mpg = np.mean(mpg[cylinders == cyl])
    print(f"{int(cyl)} cylinders: {mean_mpg:.2f}")
```

```python
# 1

df = pd.read_csv(r"C:\Users\ASUS\Downloads\auto-mpg.csv")
print(df.head(10))
print(df.shape)
print(df.describe())
```

```python
# 2

df.columns = ["mpg","cylinders","displacement","horsepower","weight","acceleration","model_year","origin","car_name"]

df["weight"] = pd.to_numeric(df["weight"], errors="coerce")
df["model_year"] = pd.to_numeric(df["model_year"], errors="coerce")

filtered = df[(df["model_year"] == 75) & (df["weight"] < 3000)]

result = filtered[["car_name", "weight", "mpg"]]

print(result)
```

```python
# 3

df.columns = ["mpg","cylinders","displacement","horsepower",
              "weight","acceleration","model_year","origin","car_name"]


df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")

print("Missing values in each column:")
print(df.isna().sum())

median_hp = df["horsepower"].median()

df["horsepower"] = df["horsepower"].fillna(median_hp)

print("\nMissing values after filling horsepower column:")
print(df.isna().sum())
```

```python
# 4

df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")
df["weight"] = pd.to_numeric(df["weight"], errors="coerce")

df["power_to_weight_ratio"] = df["horsepower"] / df["weight"]

print(df[["car_name", "horsepower", "weight", "power_to_weight_ratio"]].head())
```

```python
# 5

origin_map = {1: "USA", 2: "Europe", 3: "Japan"}
df["origin_name"] = df["origin"].map(origin_map)

mean_mpg_by_origin_named = df.groupby("origin_name")["mpg"].mean()

print("\nMean MPG by Origin (Labeled):")
print(mean_mpg_by_origin_named)
```

```python
# 6

top_10_mpg = df.sort_values(by='mpg', ascending=False).head(10)
print(top_10_mpg)
```

```python
def performance_score(row):
    return row['mpg'] * row['acceleration'] / row['weight']

df['performance_score'] = df.apply(performance_score, axis=1)
print(df[['mpg', 'acceleration', 'weight', 'performance_score']].head())
```
