# 1. Bank Account Manager
1. Simulate a bank account with deposit, withdrawal, and balance enquiry operations
2. Use variables and data types to store account information
3. Implement conditional statements to check for sufficient balance before withdrawal
4. Use functions to perform deposit, withdrawal, and balance inquiry operations.

```python
balance = 0.0

def deposit(amount):
    global balance
    if amount > 0:
        balance += amount
        print(f"Deposited ₹{amount}. New balance: ₹{balance}")
    else:
        print("Deposit amount must be positive.")

def withdraw(amount):
    global balance
    if amount <= 0:
        print("Withdrawal amount must be positive.")
    elif amount > balance:
        print("Insufficient balance! Withdrawal failed.")
    else:
        balance -= amount
        print(f"Withdrawn ₹{amount}. New balance: ₹{balance}")

def check_balance():
    print(f"Current balance: ₹{balance}")

def main():
    print("Welcome to your Bank Account")
    while True:
        print("\nSelect operation:")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            amount = float(input("Enter amount to deposit: "))
            deposit(amount)
        elif choice == "2":
            amount = float(input("Enter amount to withdraw: "))
            withdraw(amount)
        elif choice == "3":
            check_balance()
        elif choice == "4":
            print("Thank you for banking with us!")
            break
        else:
            print("Invalid choice. Please try again.")

main()
```

#2. Weather Data Analyzer
1. Analyze weather data (temperature, humidity, etc.) for a given region
2. Use variables and data types to store weather data
3. Implement conditional statements to determine weather conditions
(sunny, rainy, etc.)
4. Use functions to calculate average temperature and humidity, and print
results

```python
def average(data):
    total = 0
    for value in data:
        total += value
    return total / len(data)

def weather_condition(temp, humidity):
    if humidity > 80:
        return "Rainy"
    elif temp > 35:
        return "Hot & Sunny"
    elif temp < 20:
        return "Cold"
    elif 20 <= temp <= 30 and humidity < 60:
        return "Pleasant"
    else:
        return "Humid"

def analyze_weather(temperatures, humidities):
    avg_temp = average(temperatures)
    avg_humidity = average(humidities)

    print("\n WEATHER DATA ANALYSIS")
    print(f"Average Temperature: {avg_temp:.2f}°C")
    print(f"Average Humidity: {avg_humidity:.2f}%")
    print(f"Overall Condition: {weather_condition(avg_temp, avg_humidity)}")

def main():
    print("Weather Data Analyzer")

    days = int(input("Enter number of days to analyze: "))

    temperatures = []
    humidities = []

    for i in range(days):
        print(f"\nDay {i+1}:")
        temp = float(input("Enter temperature (°C): "))
        humidity = float(input("Enter humidity (%): "))

        temperatures.append(temp)
        humidities.append(humidity)

    analyze_weather(temperatures, humidities)

main()
```
